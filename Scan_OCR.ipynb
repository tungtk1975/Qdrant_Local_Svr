```python
import os
import PyPDF2
from PIL import Image
import pytesseract
import io
import pdf2image
import json
file_path = '/content/drive/MyDrive/164-TTTT-HCTH-01.08.2023 Gioi thieu de cu tham gia Giai thuong.pdf'
```

```python
pdf = PyPDF2.PdfReader(file_path)
# Prepare empty list to hold text results
text_results = []
# Extract images from PDF using PyPDF2
print(len(pdf.pages))
for page_num in range(len(pdf.pages)):
    page = pdf.pages[page_num]
    xObject = page['/Resources']['/XObject'].get_object()

    for obj in xObject:
        try:
            if xObject[obj]['/Subtype'] == '/Image':
                size = (xObject[obj]['/Width'], xObject[obj]['/Height'])
                data = xObject[obj].get_data()
                print("Size:", size)
                print("Data length:", len(data))
                img = Image.open(io.BytesIO(data))

                    # Perform OCR on image
                text = pytesseract.image_to_string(img, lang='vie')
                print(text)

                    # Append text result
                text_results.append(text)
        except:
            print('passed')

    # Join text from all pages
    results = ' '.join(text_results)

# Save output
new_filename = '/content/drive/MyDrive/test'
with open(new_filename + '.txt', 'w') as f:
    f.write(results)
```
